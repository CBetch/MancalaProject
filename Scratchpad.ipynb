```python
from PlayGames import PlayGames
import numpy as np
import pandas as pd
```

```python
# Random vs Random
numgames = 100
p1type = "random"
p2type = "random"
games = PlayGames(p1type, p2type, numgames)
results = games.play_games()
print(results)

df = pd.DataFrame({
    "status": results["status"],
    "p1_score": [s[0] for s in results["score"]],
    "p2_score": [s[1] for s in results["score"]],
    "num_moves": results["num_moves"]
})
df.to_csv(f"{numgames}-{p1type}VS{p2type}-Results.csv", index=False)
```

```python
# Get results from last results
p1, p2, tie = 0, 0, 0
for res in results['status']:
    if res == 1:
        p1 += 1
    elif res == 2:
        p2 += 1
    else:
        tie += 1

p1score, p2score = 0, 0
for score in results['score']:
    p1score += score[0]
    p2score += score[1]

nummoves = 0
for moves in results['num_moves']:
    nummoves += moves

print(f"p1wins: {p1} - p2wins: {p2} - ties: {tie}")
print(f"avg p1score: {p1score/numgames} - avg p2score: {p2score/numgames}")
print(f"avg game length: {nummoves/numgames}")
```
